```python
import pandas as pd  # Para manipulación de datos
import numpy as np   # Para operaciones numéricas
import matplotlib.pyplot as plt  # Para visualizaciones
import seaborn as sns  # Para visualizaciones más estilizadas
```

```python
df = pd.read_csv('/content/Taipei City Traffic Accident Sp_- traducido.csv', sep=';')  # Try using a semicolon as the delimiter.
# Adjust the delimiter according to your file's structure. Other common delimiters are tabs ('\t') and pipes ('|').
```

```python
# Renombrar columnas para facilitar el análisis
# Obtener los nombres originales de las columnas
original_columns = df.columns.tolist()

# Definir los nuevos nombres de las columnas que deseas cambiar
new_columns = ['hora_accidente', 'tipo_gestion', 'lugar', 'x', 'y']

# Reemplazar los nombres de las columnas originales con los nuevos nombres
# hasta la longitud de new_columns
for i in range(min(len(original_columns), len(new_columns))):
    original_columns[i] = new_columns[i]

# Asignar los nombres actualizados al DataFrame
df.columns = original_columns

# Verificar los nombres de las columnas
print(df.columns)

# Continuar con el resto del código
# ...
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

# Paleta de colores más suave
colores = sns.color_palette("pastel")

# Estilo general
sns.set(style="whitegrid", font_scale=1.2)

# 1. Accidentes por mes
plt.figure(figsize=(10, 5))
sns.countplot(x='mes', data=df, palette=colores)
plt.title('🚗 Accidentes por mes en Taipéi (2021)', fontsize=16, fontweight='bold')
plt.xlabel('Mes')
plt.ylabel('Número de accidentes')
plt.xticks(ticks=range(0,12), labels=["Ene", "Feb", "Mar", "Abr", "May", "Jun",
                                      "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"])
plt.tight_layout()
plt.show()

# 2. Accidentes por hora
plt.figure(figsize=(10, 5))
sns.countplot(x='hora', data=df, palette=colores)
plt.title('🕒 Accidentes por hora del día', fontsize=16, fontweight='bold')
plt.xlabel('Hora del día')
plt.ylabel('Número de accidentes')
plt.xticks(ticks=range(0, 24))
plt.tight_layout()
plt.show()

# 3. Accidentes por día de la semana
plt.figure(figsize=(10, 5))
orden_dias = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
dias_es = ['Lun', 'Mar', 'Mié', 'Jue', 'Vie', 'Sáb', 'Dom']
sns.countplot(x='dia_semana', data=df, order=orden_dias, palette=colores)
plt.title('📅 Accidentes por día de la semana', fontsize=16, fontweight='bold')
plt.xlabel('Día')
plt.ylabel('Número de accidentes')
plt.xticks(ticks=range(7), labels=dias_es)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ajustamos la figura para tener tres subgráficas
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Gráfico de distribución por hora
sns.histplot(df['hora_accidente'].dt.hour, kde=False, ax=axes[0], color="skyblue", bins=24)
axes[0].set_title('Distribución de Accidentes por Hora')
axes[0].set_xlabel('Hora del Accidente')
axes[0].set_ylabel('Número de Accidentes')

# Gráfico de distribución por tipo de gestión
sns.countplot(data=df, x='tipo_gestion', ax=axes[1], palette="Set2")
axes[1].set_title('Distribución de Accidentes por Tipo de Gestión')
axes[1].set_xlabel('Tipo de Gestión')
axes[1].set_ylabel('Número de Accidentes')

# Gráfico de distribución geográfica
sns.scatterplot(x=df['x'], y=df['y'], ax=axes[2], color='orange', alpha=0.5)
axes[2].set_title('Distribución Geográfica de Accidentes')
axes[2].set_xlabel('Coordenada X')
axes[2].set_ylabel('Coordenada Y')

# Ajustar el espacio entre los gráficos
plt.tight_layout()

# Mostrar la figura
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ajustamos la figura para tener tres subgráficas en una fila
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Gráfico de distribución por hora (estilo bonito)
sns.histplot(df['hora_accidente'].dt.hour, kde=False, ax=axes[0], color="lightseagreen", bins=24)
axes[0].set_title('Distribución de Accidentes por Hora', fontsize=14, weight='bold')
axes[0].set_xlabel('Hora del Accidente', fontsize=12)
axes[0].set_ylabel('Número de Accidentes', fontsize=12)
axes[0].tick_params(axis='both', labelsize=10)
axes[0].grid(True, linestyle='--', alpha=0.7)

# Gráfico de distribución por tipo de gestión (estilo bonito)
sns.countplot(data=df, x='tipo_gestion', ax=axes[1], palette="muted")
axes[1].set_title('Distribución de Accidentes por Tipo de Gestión', fontsize=14, weight='bold')
axes[1].set_xlabel('Tipo de Gestión', fontsize=12)
axes[1].set_ylabel('Número de Accidentes', fontsize=12)
axes[1].tick_params(axis='both', labelsize=10)
axes[1].grid(True, linestyle='--', alpha=0.7)

# Gráfico de distribución geográfica (estilo bonito)
sns.scatterplot(x=df['x'], y=df['y'], ax=axes[2], color='tomato', alpha=0.6, s=30)
axes[2].set_title('Distribución Geográfica de Accidentes', fontsize=14, weight='bold')
axes[2].set_xlabel('Coordenada X', fontsize=12)
axes[2].set_ylabel('Coordenada Y', fontsize=12)
axes[2].tick_params(axis='both', labelsize=10)
axes[2].grid(True, linestyle='--', alpha=0.7)

# Ajustar el espacio entre los gráficos
plt.tight_layout()

# Mostrar la figura
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aseguramos que 'hora_accidente' esté en formato datetime
df['hora_accidente'] = pd.to_datetime(df['hora_accidente'])

# Extraemos la hora, mes y día de la semana
df['hora'] = df['hora_accidente'].dt.hour
df['mes'] = df['hora_accidente'].dt.month
df['dia_semana'] = df['hora_accidente'].dt.day_name()

# Ajustamos la figura para tener tres subgráficas en una fila
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Gráfico de distribución por hora
sns.histplot(df['hora'], kde=False, ax=axes[0], color="lightseagreen", bins=24)
axes[0].set_title('Distribución de Accidentes por Hora', fontsize=14, weight='bold')
axes[0].set_xlabel('Hora del Accidente', fontsize=12)
axes[0].set_ylabel('Número de Accidentes', fontsize=12)
axes[0].tick_params(axis='both', labelsize=10)
axes[0].grid(True, linestyle='--', alpha=0.7)

# Gráfico de distribución por mes
sns.countplot(data=df, x='mes', ax=axes[1], palette="muted")
axes[1].set_title('Distribución de Accidentes por Mes', fontsize=14, weight='bold')
axes[1].set_xlabel('Mes', fontsize=12)
axes[1].set_ylabel('Número de Accidentes', fontsize=12)
axes[1].tick_params(axis='both', labelsize=10)
axes[1].grid(True, linestyle='--', alpha=0.7)

# Gráfico de distribución por día de la semana
sns.countplot(data=df, x='dia_semana', ax=axes[2], palette="coolwarm")
axes[2].set_title('Distribución de Accidentes por Día de la Semana', fontsize=14, weight='bold')
axes[2].set_xlabel('Día de la Semana', fontsize=12)
axes[2].set_ylabel('Número de Accidentes', fontsize=12)
axes[2].tick_params(axis='both', labelsize=10)
axes[2].grid(True, linestyle='--', alpha=0.7)

# Ajustar el espacio entre los gráficos
plt.tight_layout()

# Mostrar la figura
plt.show()
```
