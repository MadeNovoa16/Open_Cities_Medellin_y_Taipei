```python
import pandas as pd
```

```python
df_calidad_aire = pd.read_csv('/content/Taipei City Air pollution_Traducido.csv')
```

```python
df_calidad_aire
```

```python
# Verificar los valores faltantes por columna
missing_values = df_calidad_aire.isnull().sum()

# Calcular el porcentaje de valores faltantes por columna
missing_percentage = (missing_values / len(df_calidad_aire)) * 100

# Crear un DataFrame con la información de los valores faltantes
missing_data = pd.DataFrame({
    'Valores Faltantes': missing_values,
    'Porcentaje': missing_percentage
})

# Mostrar la tabla con los valores faltantes
print("Valores faltantes en cada columna:")
display(missing_data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convertir la columna de Año_Gregoriano a tipo de dato adecuado (si es necesario)
df_calidad_aire['Año_Gregoriano'] = pd.to_datetime(df_calidad_aire['Año_Gregoriano'], format='%Y')

# Gráfico de la evolución de una variable (por ejemplo, partículas suspendidas)
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_calidad_aire, x='Año_Gregoriano', y='Partículas suspendidas [μg/m3]', marker='o')
plt.title('Evolución de Partículas Suspendidas a lo largo de los Años')
plt.xlabel('Año')
plt.ylabel('Concentración de Partículas Suspendidas [μg/m3]')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convertir la columna de 'Año_Gregoriano' a tipo de dato datetime si no lo está
df_calidad_aire['Año_Gregoriano'] = pd.to_datetime(df_calidad_aire['Año_Gregoriano'], errors='coerce')

# Graficar las tendencias de los contaminantes más relevantes
plt.figure(figsize=(14, 8))

# Graficar PM2.5
plt.subplot(2, 3, 1)
sns.lineplot(x='Año_Gregoriano', y='Partículas suspendidas [μg/m3]', data=df_calidad_aire)
plt.title('Tendencia PM2.5')
plt.xlabel('Año')
plt.ylabel('Concentración (μg/m3)')

# Graficar NO2
plt.subplot(2, 3, 2)
sns.lineplot(x='Año_Gregoriano', y='Dióxido de nitrógeno [PPM]', data=df_calidad_aire)
plt.title('Tendencia NO2')
plt.xlabel('Año')
plt.ylabel('Concentración (PPM)')

# Graficar CO
plt.subplot(2, 3, 3)
sns.lineplot(x='Año_Gregoriano', y='Monóxido de carbono [PPM]', data=df_calidad_aire)
plt.title('Tendencia CO')
plt.xlabel('Año')
plt.ylabel('Concentración (PPM)')

# Graficar SO2
plt.subplot(2, 3, 4)
sns.lineplot(x='Año_Gregoriano', y='Dióxido de azufre [PPM]', data=df_calidad_aire)
plt.title('Tendencia SO2')
plt.xlabel('Año')
plt.ylabel('Concentración (PPM)')

# Graficar O3
plt.subplot(2, 3, 5)
sns.lineplot(x='Año_Gregoriano', y='Ozono [ppm]', data=df_calidad_aire)
plt.title('Tendencia O3')
plt.xlabel('Año')
plt.ylabel('Concentración (PPM)')

# Ajustar los subgráficos
plt.tight_layout()
plt.show()
```
